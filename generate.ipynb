```python
import os
from cairosvg import svg2png
import chess
import chess.svg

def generate_png_from_fen(FEN, output_dir):
    board = chess.Board(FEN)
    boardsvg = chess.svg.board(board=board, size=400, coordinates=False)
    FEN = FEN.replace('/', '_')
    svg2png(bytestring=boardsvg, write_to=os.path.join(output_dir, FEN + '.png'))

generate_png_from_fen('rnbqkbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBQKBNR', 'data')
generate_png_from_fen('rnbkqbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBKQBNR', 'data')
```

```python
# Generate random FEN
import random
import numpy as np

board = chess.Board()

# random move
for _ in range(50):
    legal_moves = list(board.legal_moves)
    random_move = random.choice(legal_moves)
    board.push(random_move)

print(board.fen())
fen_clear = board.fen().split(' ')[0]
generate_png_from_fen(fen_clear, 'data')
```

```python
# split each chess board into 64 images
import glob
import numpy as np
from PIL import Image

def split_image_into_64(img_path, output_dir):
    FEN = os.path.basename(img_path).replace('.png', '').replace('_', '/')
    print(FEN)
    board = chess.Board(FEN)

    img = Image.open(img_path)
    img = np.array(img)
    height, width, _ = img.shape
    for i in range(8):
        for j in range(8):
            # get piece symbol at (i, j)
            piece = board.piece_at(8 * (7-i) + j)
            piece_symbol = 'E' if piece is None else piece.symbol()
            print(piece_symbol, end=' ')

            file_number = 0
            file_name = f'{piece_symbol}_{file_number}.png'
            while os.path.exists(os.path.join(output_dir, file_name)):
                file_number += 1
                file_name = f'{piece_symbol}_{file_number}.png'

            x = i * height // 8
            y = j * width // 8
            cropped_img = img[x:x+height//8, y:y+width//8]
            cropped_img = Image.fromarray(cropped_img)
            cropped_img.save(os.path.join(output_dir, file_name))

    print()

# clear data_split directory
for file in glob.glob('data/data_split/*'):
    os.remove(file)
os.makedirs('data/data_split', exist_ok=True)

for img_path in glob.glob('data/*.png'):
    split_image_into_64(img_path, 'data/data_split')
```
